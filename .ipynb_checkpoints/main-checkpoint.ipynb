```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 3*x**2 + 2*x + 1
```

```python
f(3.0)
```

```python
xs = np.arange(-5, 5, 0.5)
ys = f(xs)
plt.plot(xs, ys)
plt.show()
```

```python
h = 0.0000000000000001
x = 3.0
(f(x + h) - f(x))/h
```
